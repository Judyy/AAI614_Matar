```python
df=dff
```

```python
# Create a new column, pre_tip_amount
df['pre_tip_amount'] = (
    df
    [['fare_amount', 'extra', 'mta_tax', 'tolls_amount',
      'improvement_surcharge', 'congestion_surcharge']]
    .sum(axis='columns')
)
```

```python
# Create a new column, tip_percentage, showing the percent of the fare_amount that people tipped
df['tip_percentage'] = df['tip_amount'] / df['pre_tip_amount']
```

```python
# What was the overall tip percentage?
df['tip_percentage'].mean()
```

```python
# How many times did people tip more than 100%?
(df['tip_percentage'] > 1).value_counts()
print(f"people tip  more than 100%,   {(df['tip_percentage'] > 1).value_counts()[1]} times")
```

```python
# What percent of taxi riders give no tip at all?
(df['tip_percentage'] == 0).value_counts(normalize=True)
print(f"the percentage of taxi riers that give no tip at all is {(df['tip_percentage'] == 0).value_counts(normalize=True)[1]*100:.2f}%")
```

# Questions

1. We saw that 32 percent of riders don't tip at all. Of those who *do*, what percentage do they tip, on average?

```python
# Filter out rows where tip_percentage is 0
df_tipped = df[df['tip_percentage'] > 0]

# Calculate the mean tip percentage for riders who tipped
average_tip_percentage = df_tipped['tip_percentage'].mean()

print(average_tip_percentage)
```

```python
average_tip_percentage = df[df['tip_percentage'] > 0]['tip_percentage'].mean()
print(f"Of those who tips, the percentage is {average_tip_percentage*100:.2f} %")
```

2. How many of the rides in our data set, supposedly from January and July 2019, are from outside of those dates?

```python
# #Selecting the data outside jan and july
# outside_jan_july = df[~((df.index >= '2019-01-01') & (df.index <= '2019-01-31')) &
#                       ~((df.index >= '2019-07-01') & (df.index <= '2019-07-31'))]
# num_outside = outside_jan_july.shape[0] # this is define the number of rides
# print(f"the number of rides outside the month of January and July 2019 is {num_outside} rides")
outside_dates = df[~((df.index >= '2019-01-01') & (df.index <= '2019-01-31')) &
                   ~((df.index >= '2019-07-01') & (df.index <= '2019-07-31'))]
num_outside_dates = outside_dates.shape[0]
print("Number of rides outside January and July 2019:", num_outside_dates)
```

3. Looking only at dates in January and July, on what week did passengers tip the greatest percentage?

```python
# Resample data by week ('W') and calculate the mean tip percentage for each week
weekly_tip_percentage = df.resample('W')['tip_percentage'].mean()

# Sort the weeks by tip percentage in descending order and get the top week
top_week = weekly_tip_percentage.sort_values(ascending=False).head(1)

print("Week with the highest average tip percentage:")
print(top_week)
```
